```python
Q1 = df_dropna['Average gustspeed (mph)'].quantile(0.25)
Q2 = df_dropna['Average gustspeed (mph)'].quantile(0.5)
Q3 = df_dropna['Average gustspeed (mph)'].quantile(0.75)
q1 = df_dropna['Maximum windspeed (mph)'].quantile(0.25)
q2 = df_dropna['Maximum windspeed (mph)'].quantile(0.5)
q3 = df_dropna['Maximum windspeed (mph)'].quantile(0.75)
q1s = df_dropna['Maximum gust speed (mph)'].quantile(0.25)
q2s = df_dropna['Maximum gust speed (mph)'].quantile(0.5)
q3s = df_dropna['Maximum gust speed (mph)'].quantile(0.75)
```

#IQR

```python
IQR = Q3-Q1
iqr = q3-q1
iqrs = q3s-q1s
```

```python
lower = Q1 - 1.5 *(IQR)
upper = Q3 + 1.5 * (IQR)
low = q1 - 1.5 *(iqr)
up = q3 + 1.5 * (iqr)
lows = q1s - 1.5 *(iqrs)
ups = q3s + 1.5 * (iqrs)
```

```python
print(upper)
print(up)
print(ups)
```

```python
print(lower)
print(low)
print(lows)
```

```python
index =df_dropna[(df_dropna['Average gustspeed (mph)'] < lower) | (df_dropna['Average gustspeed (mph)'] > upper)].index
indexe =df_dropna[(df_dropna['Maximum windspeed (mph)'] < low) | (df_dropna['Maximum windspeed (mph)'] > up)].index
indexs =df_dropna[(df_dropna['Maximum gust speed (mph)'] < lows) | (df_dropna['Maximum gust speed (mph)'] > ups)].index
```

### 8. Covariance

Covariance ระหว่าง Temperature and Humidity



*   การหา Covariance ระหว่างอุณหภูมิและความชื้นนั้นทำให้เห็นว่าข้อมูลเหล่านี้มีความสัมพันธ์กันมีความสัมพันธ์บวก, ลบ, หรือไม่มีความสัมพันธ์กัน





> หากค่าความแปรปรวนร่วมเป็นบวกมาก หมายความว่าเมื่ออุณหภูมิเพิ่มขึ้น ความชื้นก็จะมีแนวโน้มเพิ่มขึ้นตาม

> และถ้าค่าความแปรปรวนร่วมเป็นลบมาก หมายความว่า เมื่ออุณหภูมิเพิ่มขึ้น ความชื้นก็จะมีแนวโน้มลดลง



```python
temperature = df_dropna['Average temperature (°F)']
humidity = df_dropna['Average humidity (%)']
```

```python
covariance_temperature_humidity = np.cov(temperature, humidity)[0, 1]
```

```python
covariance_temperature_humidity
```
